```python
fuel_consumption = data['FuelConsumption']

# Создание списка пробега и добавление среднего значения 'mean', где l == 0
mileage = []
for val in fuel_consumption:
    l = val.split("/")[0]
    l = float(l.split(' ')[0])
    if l == 0: # добавляю минимум значение if l == 0
        l = 'mean'
        mileage.append(l)
    else:
        mileage.append(round(100/l))

# Получение индексов для всех средних значений 'mean' 
mean_ind = []
for i,val in enumerate(mileage):
    if val == 'mean':
        mean_ind.append(i)

# Удаление среднего значения, обусловленного малым значением L
for val in mileage:
    try:
        mileage.remove('mean')
    except:
        pass
    
# Добавление совокупных значений набора данных в индекс средних значений
for ind in mean_ind:
    mileage.insert(ind, sum(mileage)//len(mileage))
```

```python
data['Mileage'] = mileage
data.sample(3)
```

<h1> #4 hypothesys

```python
import numpy as np
from scipy import stats

N = 10
a = np.random.randn(N) + 2
b = np.random.randn(N)


var_a = a.var(ddof=1)
var_b = b.var(ddof=1)

#std deviation
s = np.sqrt((var_a + var_b)/2)
s

## Calculate the t-statistics
t = (a.mean() - b.mean())/(s*np.sqrt(2/N))

## Compare with the critical t-value
#Degrees of freedom
df = 2*N - 2
#p-value after comparison with the t 
p = 1 - stats.t.cdf(t,df=df)
print("t = " + str(t))
print("p = " + str(2*p))

## Cross Checking with the internal scipy function
t2, p2 = stats.ttest_ind(a,b)
print("t = " + str(t2))
print("p = " + str(p2))
```

```python
Nx = 70
Ny = 40
x = np.random.randn(Nx) 
y = np.random.randn(Ny)+2
print("Nx: {}, Ny: {}".format(len(x), len(y)))
t3, p3 = stats.ttest_ind(y,x)
print("t = " + str(t3))
print("p = " + str(p3))
```
